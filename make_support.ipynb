```python
import genfire as gf
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
?gf.utility.smooth3D
```

```python
arr = gf.fileio.readVolume('results.mrc')
arr.shape
```

```python
plt.figure()
plt.imshow(np.sum(arr, axis=2))
plt.show()
```

```python
smoothing_kernel_sigma = 0.05
smoothed_rec = gf.utility.smooth3D(arr, smoothing_kernel_sigma)
plt.figure()
plt.imshow(np.sum(smoothed_rec, axis=2))
plt.colorbar()
plt.show()
```

```python
threshold = 0.03
support = smoothed_rec > threshold
plt.figure()
# plt.imshow(np.sum(support, axis=2))
plt.imshow(support[:, :, support.shape[2]//2])
plt.colorbar()
plt.show()
```

```python
gf.fileio.writeVolume('tightSupport.npy', support)
```
